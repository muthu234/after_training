```python
import pygame

# Initialize pygame
pygame.init()

# Set up the display
width, height = 800, 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Graphical Text")

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Define font
font = pygame.font.Font(None, 36)

# Define text and position
text = "Hello, World!"
text_render = font.render(text, True, WHITE)
text_rect = text_render.get_rect()
text_rect.center = (width // 2, height // 2)

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill(BLACK)

    # Draw the text
    screen.blit(text_render, text_rect)

    # Update the display
    pygame.display.flip()

# Quit pygame
pygame.quit()
```

```python
import turtle

# Function to draw a star
def draw_star():
    star = turtle.Turtle()
    star.right(75)
    star.forward(100)
    for i in range(4):
        star.right(144)
        star.forward(100)
    turtle.done()

# Function to print a name with color
def print_name_with_color(name):
    colors = ["red", "green", "blue", "orange", "purple"]  # List of colors
    turtle.penup()
    turtle.goto(-200, 0)  # Starting position
    turtle.pendown()

    for char in name:
        color_index = ord(char) % len(colors)  # Calculate color index based on ASCII value of character
        turtle.color(colors[color_index])  # Set color
        turtle.write(char, font=("Arial", 24, "bold"))  # Write character
        turtle.forward(30)  # Move turtle forward

    turtle.done()

# Call function to print a name with color
print_name_with_color("John")
```
